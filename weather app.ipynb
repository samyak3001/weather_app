```python
import requests
from datetime import datetime, timezone, timedelta

# 🔐 API KEY
API_KEY = "96c73ea634856d67ace6716d27c2662e"

# 🌍 City input
CITY = input("Enter city name: ")

# 🌐 Current Weather API
URL = f"https://api.openweathermap.org/data/2.5/weather?q={CITY}&appid={API_KEY}&units=metric"

response = requests.get(URL)
data = response.json()

# ✅ Check response
if response.status_code == 200:
    
    print("\n🌤 WEATHER REPORT")
    print("---------------------------")
    
    # 📍 Location
    print(f"City: {data['name']}")
    print(f"Country: {data['sys']['country']}")
    
    # 🌡 Temperature
    print(f"Temperature: {data['main']['temp']} °C")
    print(f"Feels Like: {data['main']['feels_like']} °C")
    print(f"Min Temp: {data['main']['temp_min']} °C")
    print(f"Max Temp: {data['main']['temp_max']} °C")
    
    # ☁ Weather
    print(f"Weather: {data['weather'][0]['main']} ({data['weather'][0]['description']})")
    
    # 💧 Extra details
    print(f"Humidity: {data['main']['humidity']}%")
    print(f"Pressure: {data['main']['pressure']} hPa")
    print(f"Visibility: {data['visibility']} meters")
    
    # 🌬 Wind
    print(f"Wind Speed: {data['wind']['speed']} m/s")
    print(f"Wind Direction: {data['wind']['deg']}°")
    
    # ⏰ TIME SECTION (NO WARNING VERSION)
    timezone_offset = data["timezone"]
    
    api_time = datetime.fromtimestamp(data["dt"], timezone.utc) + timedelta(seconds=timezone_offset)
    sunrise_time = datetime.fromtimestamp(data["sys"]["sunrise"], timezone.utc) + timedelta(seconds=timezone_offset)
    sunset_time = datetime.fromtimestamp(data["sys"]["sunset"], timezone.utc) + timedelta(seconds=timezone_offset)
    
    # 💻 System time (live)
    system_time = datetime.now()
    
    print("\n⏰ TIME DETAILS")
    print("---------------------------")
    print(f"🌐 API Time (City Time): {api_time.strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"💻 System Time (Live): {system_time.strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"🌅 Sunrise: {sunrise_time.strftime('%H:%M:%S')}")
    print(f"🌇 Sunset: {sunset_time.strftime('%H:%M:%S')}")
    print(f"🌍 Timezone Offset: {timezone_offset} seconds")

else:
    print("\n❌ Error:", data.get("message", "Unable to fetch data"))
    print("\n⏰ HOURLY FORECAST (Next Hours)")
print("----------------------------------")

forecast_url = f"https://api.openweathermap.org/data/2.5/forecast?q={CITY}&appid={API_KEY}&units=metric"
forecast_res = requests.get(forecast_url)
forecast_data = forecast_res.json()

if forecast_res.status_code == 200:
    for item in forecast_data["list"][:8]:  # next ~24 hours
        time = datetime.fromtimestamp(item["dt"])
        temp = item["main"]["temp"]
        weather = item["weather"][0]["description"]
        
        print(f"{time.strftime('%d-%m %I:%M %p')} → {temp}°C, {weather}")
else:
    print("Forecast Error:", forecast_data.get("message"))
```
